Seaborn is a versatile Python data visualization library that offers a variety of plots for different types of data and analysis. Here are five types of plots that you can create using Seaborn, along with their common use cases:

Bar Plot:

Use: Compare the values of categorical data.
Common Use Cases: Comparing quantities or frequencies across different categories, showing trends in categorical data.

Histogram:

Use: Display the distribution of a continuous variable.
Common Use Cases: Identifying data distribution, spotting potential outliers, understanding the spread of data.

Scatter Plot:

Use: Show the relationship between two continuous variables.
Common Use Cases: Identifying correlations, patterns, clusters, and outliers in data, assessing the strength and direction of relationships.

Box Plot:

Use: Visualize the distribution of data, including quartiles, median, and potential outliers.
Common Use Cases: Identifying the central tendency, spread, and skewness of a dataset, detecting outliers.

Pair Plot:

Use: Display pairwise relationships in a dataset.
Common Use Cases: Visualizing relationships between multiple variables in one plot, assessing correlations and patterns across variables.


```python
#Q2
import seaborn as sns
fmri = sns.load_dataset("fmri")
fmri
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
sns.lineplot(x = "timepoint", y = "signal", data = fmri, style = "event", hue = "region")
plt.title('signal-time')
plt.show()
```

```python
#Q3
titanic = sns.load_dataset("titanic")
```

```python
fig, axes = plt.subplots(1,2, figsize=(10, 5))
sns.boxplot(data = titanic, x = "pclass", y = "age", ax = axes[0])
axes[0].set_title(" class-age")

sns.boxplot(data = titanic, x = "pclass", y = "fare", ax = axes[1])
axes[1].set_title("class-fare")
```

```python
#Q4
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
plt.figure(figsize=(5,5))
sns.histplot(data = diamonds, x = "price", hue = "cut")
plt.title("price-dist")
plt.show()
```

```python
#Q5
iris = sns.load_dataset("iris")
iris
```

```python
sns.pairplot(data = iris, hue = "species")
```

```python
#Q6
flights = sns.load_dataset("flights")
flights
```

```python
flights_matrix = flights.pivot_table(index="year", columns="month", values="passengers")
flights_matrix
```

```python
sns.heatmap(data = flights_matrix, 
            annot = True, 
            fmt = "d",
            linewidths = 0.5,
            linecolor = "pink")
plt.title("passengers-1949-1960")
plt.show()
```
